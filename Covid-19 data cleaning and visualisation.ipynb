## explore the effect of movement control order in Covid control  

```python
mco1_date = df_mys['date'][(df_mys['date'] == '2020-03-18')]
cmco_date = df_mys['date'][(df_mys['date'] == '2020-05-04')]
rmco_date = df_mys['date'][(df_mys['date'] == '2020-06-10')]
emco_date = df_mys['date'][(df_mys['date']== '2020-09-08')]
mco2_date = df_mys['date'][(df_mys['date'] == '2021-01-12')]
```

```python
# create a lineplot of daily new cases in Malaysia against date (from 2020-01-01 to 2021-05-03)
fig, ax = plt.subplots(1,1, figsize = (24, 8))
#plt.show = sns.lineplot(x='date', y='aged_65_older', data = df_mys, ax= ax, alpha = 0.4, color = 'skyblue')
plt.show = sns.lineplot(x='date', y='new_cases', data = df_mys, ax= ax, alpha = 0.6, color='skyblue')
plt.show = sns.lineplot(x='date', y='new_cases_smoothed', data = df_mys, ax= ax, linewidth = 2.5, color= '#1f77b4')

ax2 = ax.twinx()
plt.show = sns.lineplot(x='date', y='reproduction_rate', data = df_mys, ax= ax2, linewidth = 2.5, color= '#1f77b4', alpha=0.3)
# set axis limit
ax.set_ylim(0,6200)
ax.set_xlim(df_mys['date'].min(), df_mys['date'].max())

# reset xticks interval and xticklabels
ax.set_xticks(df_interval)
plt.show = ax.set_xticklabels(df_interval.strftime('%Y-%m-%d'))

# reset x and y label
ax.set_ylabel('daily_new_cases', fontsize = 12, labelpad = 25)
ax.set_xlabel('date_by_month', fontsize = 12, labelpad = 15)

# set title
fd1 = {'family': 'serif', 'color': 'black', 'size': '18'}
plt.show = plt.title('Daily New Cases in Malaysia from Jan 2020 to May 2021', fontdict = fd1)

from textwrap import fill
legend_list = ['new_daily_cases', 'new_daily_cases_smoothed', 'stringency_index']
plt.show = fig.legend(labels = [fill(i, 16) for i in legend_list], bbox_to_anchor=(0.212,0.85), title = 'legend', fontsize=10, title_fontsize= 12)

def MCO_lines(ax, y_value, text_opacity):
    # mark MCO dates with axvlines
    MCO1 = ax.axvline(mco1_date, color='darkred', linestyle='--',linewidth=2)
    CMCO = ax.axvline(cmco_date, color='darkred', linestyle='--',linewidth=2)
    RMCO = ax.axvline(rmco_date, color='darkred', linestyle='--',linewidth=2)
    EMCO = ax.axvline(emco_date, color='darkred', linestyle='--',linewidth=2)
    MCO2 = ax.axvline(mco2_date, color='darkred', linestyle='--',linewidth=2)

    # annotate MCO with axvlines
    from matplotlib.pyplot import text
    from datetime import timedelta, date
    plt.show = text(mco1_date + timedelta(days=-5), y_value, "movement control order (MCO)" , rotation=90, verticalalignment='center', alpha = text_opacity)
    plt.show =text(cmco_date + timedelta(days=-5), y_value, "conditional movement control order (CMCO)" , rotation=90, verticalalignment='center', alpha = text_opacity)
    plt.show =text(rmco_date + timedelta(days=-5), y_value, "recovery movement control order (RMCO)" , rotation=90, verticalalignment='center', alpha = text_opacity)
    plt.show =text(emco_date + timedelta(days=-5), y_value, "enhanced movement control order (EMCO)" , rotation=90, verticalalignment='center', alpha = text_opacity)
    plt.show =text(mco2_date + timedelta(days=-5), y_value, "state based movement control order (MCO 2.0)" , rotation=90, verticalalignment='center', alpha = text_opacity)
    
MCO_lines(ax, 3000, 1)
```
